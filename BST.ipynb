# 이진트리

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.parent = None
        self.left:Node = None
        self.right:Node = Noneㅋㅋ

    def preorder(self):
        """M-L-R"""
        if self is not None:
            yield self.key
        if self.left is not None:
            yield from self.left.preorder()
            
        if self.right is not None:
            yield from self.right.preorder()
    
    def inorder(self):
        """L-M-R"""
        if self.left is not None:
            yield from self.left.inorder()
        if self is not None:
            yield self.key
        if self.right is not None:
            yield from self.right.inorder()
            
    def postorder(self):
        """L-R-M"""
        if self.left is not None:
            yield from self.left.postorder()
        if self.right is not None:
            yield from self.right.postorder()
        if self is not None:
            yield self.key
    
    def __iter__(self):
        return self.preorder()
        
    def __str__(self):
        return str(f"Node: {self.key}")
```

```python
# """
#         10
#        /  \
#       5    20
#      / \   / \
#     3   7 15 30
# """

root = Node(10)

root.left = Node(5)
root.right = Node(20)

root.left.parent = root
root.right.parent = root

root.left.left = Node(3)
root.left.right = Node(7)

root.left.left.parent = root.left
root.left.right.parent = root.left

root.right.left = Node(15)
root.right.right = Node(30)

root.right.left.parent = root.right
root.right.right.parent = root.right
```

```python
for x in root.__iter__():
    print(x)
```

# 이진 탐색 트리
- 각 노드의 왼쪽 SubTree의 Key값은 노드 Key값보다 작거나 같아야 함
- 각 노드의 오른쪽 SubTree의 Key값은 노드 Key값보다 커야한다.
- 모든 노드가 위 조건을 만족해야한다.

```python
class BinarySearchTree:
    def __init__(self):
        self.root:Node = None
        self.size = 0
        
    def __len__(self):
        return self.size
    
    def __iter__(self):
        """BST 순회 (preorder방식)"""
        return self.root.__iter__() 
    
    def find_loc(self, value):
        """value 값 노드가 있다면 해당 노드 Return, 없다면 삽입될 부모노드 리턴"""
        if self.size == 0: # 노드가 없을 경우
            return None 
        
        p = None
        v:Node = self.root
        
        while v is not None:
            if v.key == value:
                return v
            elif v.key > value:
                p = v
                v = v.left
            else:
                p = v
                v = v.right
        return p
                    
    def search(self, value):
        """
        루트 노드의 Key값부터 비교
        
        1) value < key -> 오른쪽으로 이동
        
        2) value > key -> 왼쪽으로 이동

        3) Key값 있으면 반환, 없으면 None
        """
        if self.size == 0: # 노드가 없을 경우
            return None
        
        v:Node = self.root
        while v is not None:
            if v.key == value:
                return v
            elif v.key > value:
                v = v.left
            else:
                v = v.right
        return None
    
    def insert(self, value):
        p = None
        v:Node = self.root
        new_node = Node(value)
        
        if self.size == 0: # 노드가 하나도 없으면,
            self.root = new_node
            self.size += 1 
            return 
        
        while v:
            if v.key > value:
                p = v
                v = v.left
            elif v.key < value:
                p = v
                v = v.right
            else:
                return "중복 안됨"
        
        new_node.parent = p
        
        if p.key <= value:
            p.right = new_node
        
        else:
            p.left = new_node
        
        self.size += 1
        return new_node
    
    def delete(self, value):
        pt:Node = self.root # 부모 노드
        m:Node = None # 삭제 노드의 왼쪽 서브트리 중 가장 높은 값을 가진 노드
        c:Node = None # 삭제할 노드 자리를 대신할 노드
        v:Node = self.root # 현재 노드
        # 삭제할 노드 탐색
        while v:
            if v.key < value:
                pt = v
                v = v.right
            elif v.key > value:
                pt = v
                v = v.left
            else: # 해당 노드를 찾음
                break
        else:
            print("해당 노드를 찾지 못했습니다.")
            return False # 찾지 못했습니다.
        
        # pt와 v를 위에서 찾게됨
        # 삭제 연산 실행
        
        # 삭제할 노드의 자식 노드가 2개인 경우, 아래 그림에서 50을 삭제하는 경우
        #      50
        #     /  \
        #   25    75
        #           \
        #            200
        if v.left and v.right: # 삭제할 노드의 자식 노드가 2개인 경우
            c = v.left # c 찾기
            m = v.left # m 찾기
            while m.right:
                m = m.right
            m.right = v.right # 삭제 노드의 왼쪽 트리 중 가장 높은 값의 노드의 오른쪽 노도는 삭제 노드의 오른쪽 노드를 참조
            v.right.parent = m # 삭제 노드의 오른쪽 자식 노드의 부모는 M을 참조
            if v is self.root:
                self.root = c
                c.parent = None
            else:
                if pt.key > value:
                    pt.left = c
                    c.parent = pt
                elif pt.key < value:
                    pt.right = c
                    c.parent = pt
            self.size -= 1
            return True
        # 삭제할 노드가 0개 또는 1개
        if v.left:
            child = v.left
        elif v.right:
            child = v.right
        
        if v is self.root:
            self.root = child

            if child:
                child.parent = None

        elif pt.left is v:
            pt.left = child

            if child:
                child.parent = pt
        
        elif pt.right is v:
            pt.right = child

            if child:
                child.parent = pt
        
        self.size -= 1
```

```python
bst = BinarySearchTree()
bst.insert(7)
bst.insert(5)
bst.insert(12)
bst.insert(1)
bst.insert(6)
bst.insert(9)
bst.insert(14)
bst.insert(4)

for x in bst.__iter__():
    print(x)
```
