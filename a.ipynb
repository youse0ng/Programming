```python
from abc import ABC, abstractmethod
from typing import Sequence
# HEAP 성질
# 모든 부모노드의 Key값은 자식노드의 Key값보다 작지 않다. 

class HeapBase(ABC):
    """이진 트리의 모양을 만족하면서, HEAP 성질을 만족하는 자료구조"""
    @abstractmethod
    def __init__(self, sequence):
        pass
    
    @abstractmethod
    def _make_heap(self):
        pass
    
    @abstractmethod
    def _heapify_down(self, k:int, n:int):
        """맨 오른쪽부터 루트 노드까지 Heapify_down"""
        pass
    
    @abstractmethod
    def insert(self, value):
        pass
    
    @abstractmethod
    def _heapify_up(self):
        pass
    
    @abstractmethod
    def find_max(self):
        pass
    
    @abstractmethod
    def delete_max(self):
        pass
```

```python
class Heap(HeapBase):
    def __init__(self, sequence:list):
        self.sequence = sequence
    
    def __str__(self):
        self._make_heap()
        return f"HEAP: {self.sequence}"
        
    def _make_heap(self):
        """O(nlogn)"""
        n = len(self.sequence)
        for k in range(n-1,-1,-1):
            self._heapify_down(k,n)
                 
    def _heapify_down(self, k:int, n:int)->None:
        """
        맨 아래 오른쪽 리프 노드부터 루트 노드까지 Heapify_Down
        
        k: 지금 현재 주목된 k번째 노드
        
        n: 전체 노드 수
        """
        while True:
            largest_index = k
            left, right = (k*2) + 1, (k*2) + 2 # 왼쪽 노드, 오른쪽 노드
            if left < n and self.sequence[left] > self.sequence[largest_index]:
                largest_index = left
            
            if right < n and self.sequence[right] > self.sequence[largest_index]:
                largest_index = right    
            
            if largest_index == k:
                break
            
            self.sequence[k], self.sequence[largest_index] = self.sequence[largest_index], self.sequence[k]
            k = largest_index
    
    def _heapify_up(self):
        k = len(self.sequence) - 1 # 현재 입력된 노드 위치
        while k > 0 and self.sequence[k] > self.sequence[(k-1)//2]:
            self.sequence[k], self.sequence[(k-1)//2] = self.sequence[(k-1)//2], self.sequence[k]
            k = (k-1) // 2
    
    def insert(self, value):
        self.sequence.append(value)
        self._heapify_up()
        print(f"value_insert:{value}\n heap: {self.sequence}")

    def find_max(self):
        return self.sequence[0]
    
    def delete_max(self):
        # 최대 값 저장
        max_value = self.sequence[0]
        
        # 마지막 리프 노드를 루트 노드로 이동
        last_node_index = len(self.sequence) - 1
        self.sequence[0] = self.sequence[last_node_index]
        
        # 마지막 노드 삭제
        self.sequence.pop()
        
        # Heapify_down을 루트 노드 부터 실행
        self._heapify_down(0, len(self.sequence))
        print(f"{max_value} node delete \n Deleted_Heap: {self.sequence}")
        return max_value
a = Heap([15,12,6,11,10,2,3,1,8])
a.insert(14)
a.delete_max()
```
