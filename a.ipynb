```python
from abc import ABC, abstractmethod
from typing import Sequence
# HEAP 성질
# 모든 부모노드의 Key값은 자식노드의 Key값보다 작지 않다. 

class HeapBase(ABC):
    """이진 트리의 모양을 만족하면서, HEAP 성질을 만족하는 자료구조"""
    @abstractmethod
    def __init__(self, sequence):
        pass
    
    @abstractmethod
    def _make_heap(self):
        pass
    
    @abstractmethod
    def _heapify_down(self, k:int, n:int):
        """맨 오른쪽부터 루트 노드까지 Heapify_down"""
        pass
    
    @abstractmethod
    def insert(self, value):
        pass
    
    @abstractmethod
    def find_max(self):
        pass
    
    @abstractmethod
    def delete_max(self):
        pass
```

```python
class Heap(HeapBase):
    def __init__(self, sequence:list):
        self.sequence = sequence
        
    def _make_heap(self):
        """O(nlogn)"""
        n = len(self.sequence)
        for k in range(n-1,-1,-1):
            self._heapify_down(k,n)
                 
    def _heapify_down(self, k:int, n:int)->None:
        """맨 아래 오른쪽 리프 노드부터 루트 노드까지 Heapify_Down
        k: 지금 현재 주목된 k번째 노드
        n: 전체 노드 수
        """
        while True:
            largest_index = k
            left, right = (k*2) + 1, (k*2) + 2 # 왼쪽 노드, 오른쪽 노드
            if left < n and self.sequence[left] > self.sequence[largest_index]:
                largest_index = left
            
            if right < n and self.sequence[right] > self.sequence[largest_index]:
                largest_index = right    
            
            if largest_index == k:
                break
            
            self.sequence[k], self.sequence[largest_index] = self.sequence[largest_index], self.sequence[k]
            k = largest_index 
    def __str__(self):
        self._make_heap()
        return f"HEAP: {self.sequence}"
        
a = Heap([3,4,6,1,2,11,10])
print(a)
```
